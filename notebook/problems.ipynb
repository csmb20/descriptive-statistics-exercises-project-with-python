## Exercise 1 – How long are movie titles?

In this exercise, you will apply descriptive statistics techniques to a real dataset: a database of 1000 popular movies from the IMDb website.

From the title of each movie, you will:
- Calculate the title length (in characters).
- Obtain key statistical metrics: mean, median, mode, range, variance, standard deviation, skewness, and kurtosis.
- Identify the longest and shortest titles.
- Visualize the distribution of these lengths with a histogram and a density curve.


```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
from scipy.stats import skew, kurtosis, mode


# Load IMDb dataset
url = "../assets/imdb_1000.csv"
df = pd.read_csv(url)
df
```

```python
# Create a new column called 'title_length' with the length (number of characters) of each title

df['title_length'] = df['title'].apply(lambda x: len(x.replace(" ", ""))) # This omits blank spaces in the titles
df
```

```python
# Calculate mean, median, and mode of the 'title_length' column

# Mean
meanTitleLen   = df['title_length'].mean()
print(f'The mean of the movie title character length is {meanTitleLen:0.2f}')

# Median
medianTitleLen = df['title_length'].median()
print(f'The median of the movie title character length is {medianTitleLen:0.2f}')

# Mode
modeTitleLen  = sts.mode(df['title_length'])
print(f'The mode of the movie title character length is {modeTitleLen[1]} where most repeated value is {modeTitleLen[0]}')
```

```python
# Calculate the range, variance, and standard deviation of 'title_length'

# Standard deviation
stDevTitleLen = df['title_length'].std()
print(f'The standard deviation of the movie title character length is {stDevTitleLen:0.2f}')

# Variance
varTitleLen   = stDevTitleLen**2
print(f'The variance of the movie title character length is {varTitleLen:0.2f}')

# Range
rangeTitleLen = df['title_length'].max() - df['title_length'].min()
print(f'The range of the movie title character length is {rangeTitleLen:0.2f}')
```

```python
# Calculate skewness and kurtosis of the 'title_length' column

# Skewness
skwTitleLen     = sts.skew(df['title_length'])
print(f'The skewness of the movie title character length is {skwTitleLen:0.2f} (positive skew)')

# Kurtosis
kurtTitleLen    = sts.kurtosis(df['title_length'])
print(f'The kurtosis of the movie title character length is {kurtTitleLen:0.2f}')

# Excess kurtosis
excKurtTitleLen = 3 - kurtTitleLen
print(f'The excess kurtosis of the movie title character length is {excKurtTitleLen:0.2f} (slightly platykurtic distribution)')
```

```python
# Print the shortest and longest title based on their length

# Print the shortest title
shortestTitleLen = df[df['title_length']==df['title_length'].min()]['title'].item()
print('The shortest movie title is: ' + '\"' + shortestTitleLen + '\"')

# Print the longest title
longestTitleLen = df[df['title_length']==df['title_length'].max()]['title'].item()
print('The longest movie title is: ' + '\"' + longestTitleLen + '\"')
```

```python
# plot your results

fig, ax = plt.subplots(nrows=1,ncols=1)
ax.hist(df['title_length'],12,color='red')
ax.plot([meanTitleLen,meanTitleLen],[0,300],c='black',linestyle='dotted',label='mean')
ax.plot([medianTitleLen,medianTitleLen],[0,300],c='blue',linestyle='dashed',label='median')
ax.plot([modeTitleLen[0],modeTitleLen[0]],[0,300],c='green',linestyle='solid',label='mode')
ax.set_xlabel('Title character length')
ax.set_ylabel('Frequency')
ax.set_title('Title character length histogram')
plt.legend()
plt.grid()
plt.show()
```

The histogram shows the positive skewness obtained in the previous calculations. Furthermore, since the distribution is not symmetric, the mode, the median, and the mean are not perfectly aligned. In a positive skewed distribution mean(x) > median(x) > mode(x)

## Exercise 2 – Manual Calculation of Standard Deviation

In this exercise, you will manually calculate the standard deviation by selecting five actual titles from the movie dataset you loaded earlier.


```python
# Create a list with 5 actual values from df['title_length'], for example: [10, 13, 14, 18, 22]

rng = np.random.seed(10)
titleLenList = []
for i in range(4):
    pickElement = np.random.choice(df['title_length'])
    titleLenList.append(pickElement)

titleLenList
```

```python
# Calculate the mean of the 5 values
meanList = np.mean(titleLenList)
meanList
```

```python
# Calculate the squared differences with respect to the mean
sqrDifList = np.sum((titleLenList - meanList)**2)
DifList    = (titleLenList - meanList)
```

```python
# Calculate the variance (average of the squared differences) and its square root to obtain the standard deviation
# Variance
varLenList = np.dot(DifList,DifList)/len(DifList)
print(f'The variance of the subset of characters lengths is {varLenList}')

#Standard deviation
stdLenList = np.sqrt(varLenList)
print(f'The standard deviation of the subset of characters lengths is {stdLenList:0.2f}')
```
